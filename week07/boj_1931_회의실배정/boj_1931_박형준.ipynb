```python
import sys
input = sys.stdin.readline

n = int(input())

meet_time = []
for _ in range(n):
    a, b = map(int, input().split())
    meet_time.append((a, b))

# 끝나는 시간 오름차순 정렬, 끝나는 시간이 같으면 시작시간 오름차순 정렬
meet_time.sort(key=lambda x:(x[1], x[0]))

# 회의실 최대로 쓰는 리스트
room = []

# n = 0 : 빈 리스트
if n == 0:
    room = []

# n = 1 : 1개
elif n == 1:
    room.append(meet_time[0])

# n = 2 : 회의 종료시간이 빠른 회의 기준
# meet_time 리스트에서 그 다음 회의의 시작시간이 끝나는 시간보다 앞서는 지 확인
elif n == 2:
    # 다음 회의 시작시간이 앞선 회의 종료시간보다 앞서지 않으면 둘 다 회의 가능
    if meet_time[0][1] <= meet_time[1][0]:
        room.append(meet_time[0])
        room.append(meet_time[1])
    # 시간이 엇갈리면 하나만 가능
    else:
        room.append(meet_time[0])

# n >= 3 : 회의 종료시간이 빠른 회의 순서대로
# meet_time 리스트에서 그 다음 회의의 시작시간이 끝나는 시간보다 앞서는 지 확인
else:
    # 첫 번째 회의는 회의실 리스트에 미리 추가
    room.append(meet_time[0])

    while True:
        if len(meet_time) == 1:
            break

        else:
            # meet_time 리스트의 첫 번째 회의의 종료시간보다
            # 두 번째 회의 시작시간이 늦으면 두 번째 회의는 room 리스트에 추가
            # 첫 번째 회의는 meet_time 리스트에서 삭제
            # 두 번째 회의 종료 시간 기준으로 그 다음 회의 시작시간과 다시 비교
            if meet_time[0][1] <= meet_time[1][0]:
                room.append(meet_time[1])
                del meet_time[0]
            # meet_time 리스트의 첫 번째 회의 종료시간보다
            # 두 번째 회의 시작시간이 빠르면 두 번째 회의는 불가능하니까 meet_time 리스트에서 삭제
            else:
                del meet_time[1]

print(len(room))
```
